## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df.drop(["EIN"], axis =1, inplace=True)
application_df.head()
```

```python
# Identify the name and counts of the companys
name_counts = application_df['NAME'].value_counts()
name_counts
```

```python
#Replace names with smaller values into 'OTHER' category
name_to_replace = list(name_counts[name_counts < 50].index)
for name in name_to_replace:
  application_df['NAME'] = application_df['NAME'].replace(name, 'Other')

# check to see if replacements was successful
application_df['NAME'].value_counts()
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
#Identify the counts for the income_amt
application_df['INCOME_AMT'].value_counts()
```

```python
#Change 'INCOME_AMT' to numerical value in order, to help better preprocess the data you feed into your model
# Conversiton table
application_df.loc[application_df['INCOME_AMT'] == '0', 'INCOME_AMT'] = 0
application_df.loc[application_df['INCOME_AMT'] == '1-9999', 'INCOME_AMT'] = 1
application_df.loc[application_df['INCOME_AMT'] == '10000-24999', 'INCOME_AMT'] = 2
application_df.loc[application_df['INCOME_AMT'] == '25000-99999', 'INCOME_AMT'] = 3
application_df.loc[application_df['INCOME_AMT'] == '100000-499999', 'INCOME_AMT'] = 4
application_df.loc[application_df['INCOME_AMT'] == '1M-5M', 'INCOME_AMT'] = 5
application_df.loc[application_df['INCOME_AMT'] == '5M-10M', 'INCOME_AMT'] = 6
application_df.loc[application_df['INCOME_AMT'] == '10M-50M', 'INCOME_AMT'] = 7
application_df.loc[application_df['INCOME_AMT'] == '50M+', 'INCOME_AMT'] = 8

#Convert the income_amt column to an integer
application_df['INCOME_AMT'] = application_df['INCOME_AMT'].astype(int)
application_df['INCOME_AMT'].value_counts()
application_df.dtypes
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
application_counts = application_df['APPLICATION_TYPE'].value_counts()
application_counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace` (CUT OFF IS BELOW 528)
application_types_to_replace = application_counts[application_counts < 528 ].index


# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure replacement was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
classification_counts = application_df['CLASSIFICATION'].value_counts()
classification_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classfication_greaterthan_1 = classification_counts[classification_counts > 1 ]
classfication_greaterthan_1
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = classification_counts[classification_counts < 1000 ].index
classifications_to_replace

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure replacement was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Check the types of all the columns before converting the categorical columns
application_df.dtypes
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
#application_dummies_df = pd.get_dummies(application_df)
categorical_cols = application_df.select_dtypes(include=['object']).columns
application_dummies_df = pd.get_dummies(application_df, columns=categorical_cols)

application_dummies_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y= application_dummies_df["IS_SUCCESSFUL"].values
X= application_dummies_df.drop(columns ="IS_SUCCESSFUL").values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train_scaled[0])
hidden_nodes_layer1 = number_input_features
hidden_nodes_layer2 = number_input_features
hidden_nodes_layer3 = number_input_features

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu")
)
# Second hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu")
)

# Third hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer3, activation="relu")
)

# Output layer
nn.add(
    tf.keras.layers.Dense(units=1, activation="sigmoid")
)
# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss = "binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Fit the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy= nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
